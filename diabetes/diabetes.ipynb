```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("diabetes.csv")
```

```python
dependent_variable = data.Outcome.values        #Sonuçları içeren değişken
independent_variable = data.drop(["Outcome"], axis =1)     #Sonuçlar hariç diğer verileri içeren değişken
```

```python
#Normalizasyon yapıldı
main_independent_variable = (independent_variable - np.min(independent_variable)) / ((np.max(independent_variable) - np.min(independent_variable)))  
#print(independent_variable.head())         #Normalizasyon öncesi veriler
#print(main_independent_variable.head())    #Normalizasyon sonrası veriler
```

```python
#Train değişkenler,sistemin sağlıklı insan  ile hasta insanı ayırt etmek için kullanılır
#test değişkenler,makine öğrenme modelimizi doğru bir şekilde çalışıyor mu diye kullanılır
x_train,y_train,x_test,y_test = train_test_split(main_independent_variable, dependent_variable, test_size = 0.2, random_state = 1)
```

```python
#KNN MODEl
knn = KNeighborsClassifier(n_neighbors = 3)   #Komşu sayısı belirlenilir
knn.fit(x_train, y_train)                     #Model eğitilir
prediction = knn.predict(x_test)              #Verileri test eder
print(knn.score(x_test,y_test))               #k = sayı için test verilerinin doğruluk oranını verir
```
